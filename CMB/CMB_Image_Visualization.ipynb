## Funciton for getting the image by id in numpy format

```python
def get_image(id):
    file_nane = "data/comm18_cut_masked_id" + str(id) + ".txt"
    im = pd.read_csv(file_nane, sep=" ", header=None).to_numpy()
    X = im[:,[0,1]]
    y = im[:,2]
    return X, y
```

## Check whether there are 128*128 elements in each image

```python
s = 0
for id in range(1034):
    X, y = get_image(id)
    if(X.shape[0] != 128*128):
        s = s + 1
s
```

## Function for displaying image for a specified id

```python
def displayImage(id):
    file_nane = "data/comm18_cut_masked_id" + str(id) + ".txt"
    im = pd.read_csv(file_nane, sep=" ", header=None).to_numpy()
    y = im[:,2]
    sns.heatmap(y.reshape(128,128), vmin=-750, vmax=500)

displayImage(0)
```
